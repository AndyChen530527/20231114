```python
import pandas as pd

scores = [
    {'chinese':78,'english':80,'math':92,'history':75,'discover':85},
    {'chinese':92,'english':60,'math':85,'history':62,'discover':73}
]
dataFrame = pd.DataFrame(index=["robert","jenny"],data=scores)
display(dataFrame.T)
```

```python
import matplotlib.pyplot as plt
figure = plt.figure()
axes = figure.add_subplot(1,1,1)
dataFrame.T.plot(ax=axes)
axes.set_title('Scores')
axes.grid(visible=True,axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
figure = plt.figure()
axes = figure.add_subplot(1,1,1)
dataFrame.T.plot(ax=axes,style=['--','-'], marker = 'o')
axes.set_title('Scores')
axes.grid(visible=True,axis='y')
plt.show()
```

```python
import pandas as pd

scores = [
    {'chinese':78,'english':80,'math':92,'history':75,'discover':85},
    {'chinese':92,'english':60,'math':85,'history':62,'discover':73}
]
dataFrame = pd.DataFrame(index=["",""],data=scores)
display(dataFrame.T)
```
